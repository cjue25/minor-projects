#### 3. Deal with the Json columns and split the contents
#### Rule:

--------------------------------------------------------------------------
I take first few terms to see the main contribution and avoid too many nan values.
- **company, country, director, producer, writer** : Take the first name
- **casts**  : Take the first three names
- **crews**  : Take the first name of Director, Writer and Producer
--------------------------------------------------------------------------
- **spoken_language** : Ignored, because the **original_language** might be enough for me
- **genres**    :  Split by |  and generate all the individual columns
- **keywords**  :  Split by | 

```python
# learn from Ref(1) that a good way to extract what I want in json style
def safe_access(container, index_values):
    # return missing value rather than an error upon indexing/key failure
    result = container
    try:
        for idx in index_values:
            result = result[idx]
        return result
    except IndexError or KeyError:
        return pd.np.nan

def get_dir(crew_data):
    gj = [x['name'] for x in crew_data if x['job'] == 'Director']
    return safe_access(gj, [0])

def get_writer(crew_data):
    gj = [x['name'] for x in crew_data if x['job'] == 'Writer']
    return safe_access(gj, [0])

def get_producer(crew_data):
    gj = [x['name'] for x in crew_data if x['job'] == 'Producer']
    return safe_access(gj, [0])

def pipe_split(column):
    return '|'.join([x['name'] for x in column])

def unique_set_list(column):
    # find unique list of the items
    uni_list=set()
    for each_row in df[column].str.split('|').values:
        if isinstance(each_row,float): continue # avoid word = NaN
        uni_list=uni_list.union(each_row)  
#     uni_list.remove('')
    return uni_list 
```

```python
# add new columns
df['country']=df.production_countries.apply(lambda x: safe_access(x, [0, 'name']))
df['company']=df.production_companies.apply(lambda x: safe_access(x, [0, 'name']))
df['director']=df['crew'].apply(get_dir)
df['producer']=df['crew'].apply(get_producer)
df['writer']=df['crew'].apply(get_writer)

df['actor1']=df['cast'].apply(lambda x: safe_access(x, [0, 'name']))
df['actor2']=df['cast'].apply(lambda x: safe_access(x, [1, 'name']))
df['actor3']=df['cast'].apply(lambda x: safe_access(x, [2, 'name']))

df['keywords_split']=df['keywords'].apply(pipe_split)
df['genres_split']=df['genres'].apply(pipe_split)

genres_list=list(unique_set_list('genres_split'))
keywords_list=unique_set_list('keywords_split')

df.head(1)
```

```python
# generate every genres columns with zero-value
df[genres_list]=pd.DataFrame([np.zeros((len(genres_list)))],index=df.index)
df.head(1)
```

#### 6. Clean missing and unreasonable value in ``revenue`` , ``budget``, `vote_count` to help the following analysis

- Define a new variable: **ratio = revenue / budget**
    
    Although I can directly compare other variables with revenue to explore the **successful** property of the films, the higher revenue should not be the successful one if its budget is high also. Thus, I define a **ratio** variable that can be another criterion to examine whether this movie is **successful** for earning money.

    On the other hand, I will compare the relation between the popularity and the vote_average based on the ratio and so on. Therefore, the zero-revenue, small budget (<100 should be wrong record), zero-vote_count will all be removed.

```python
# visulize the correlation between budget and revenue
temp=dn.copy();temp['intercept']=1
lm = sm.OLS(temp['revenue'], temp[['budget', 'intercept']])
results = lm.fit().rsquared
sb.regplot(dn.budget,dn.revenue,scatter_kws={"color": "blue"}, line_kws={"color": "red"});
plt.xlabel('budget');plt.ylabel('revenue')
plt.title('budget vs revenue (positive correlation '+r'$R^2$={:.2f})'.format(results),fontsize=15,y=1.08);
```

```python
# construct dm (Besides budget, revenue and vote_count, we remove one non-value genres )
dm=dn.copy()
dm.drop(dm.query('budget<=100').index,inplace=True)
dm.drop(dm.query('revenue==0').index,inplace=True)
dm.drop(dm.query('vote_count==0').index,inplace=True)
dm.drop(dm.query('genres==""').index,inplace=True) # only one row
```

#### 2. The percentage of quantitive variables with time

```python
def change_year(df):
    df.index=[str(x+1900) for x in df.index.tolist()]
    return df

def pie_split(ax,df,topn,other_name='others',legend=False):
    pie_df=df.sort_values(ascending=False)
    df1=pie_df[:topn].copy()
    df2=pie_df[topn:].sum()
    df1[other_name]=df2
    cs=cm.Pastel1(np.arange(9)/9.)
    if legend:
        patches, texts, autotexts=ax.pie(df1,autopct='%1.1f%%',colors=cs,startangle=0,
                                     wedgeprops = { "edgecolor" : 'k'},
                                     textprops = {"fontsize" : 15});
        ax.legend(df1.index,bbox_to_anchor=(1.05,0.8),fontsize=15)
    else:
        patches, texts, autotexts=ax.pie(df1,labels=df1.index,autopct='%1.1f%%',colors=cs,startangle=0,
                                     wedgeprops = { "edgecolor" : 'k'},
                                     textprops = {"fontsize" : 15});        
    return patches, texts, autotexts,ax
```

#### 5. The relation within multiple variables

```python
gr=[];gr2=[];new_genres_list=[]
w=0.3
for n,i in enumerate(Genres_sort.index):
    try:
        temp=dm.groupby(i)['vote_average'].median()[1]
        temp2=dm.groupby(i)['popularity'].median()[1]

    except:
        temp=0;temp2=0
        
    new_genres_list.append('('+str(n+1)+') '+i)
    gr.append(temp)
    gr2.append(temp2)
    
x=np.arange(len(Genres_sort.index))
plt.figure(figsize=(12,5))
ax1=plt.subplot(1,1,1)
plt.xticks(x+w/2,new_genres_list,rotation=90);

vote=ax1.bar(x,gr,color=(0.6,0.3,0.6),width=w,align='center')

ax2=ax1.twinx()
pop=ax2.bar(x+w,gr2,color='orange',width=w,align='center')

mean_v=ax1.axhline(y=dm.vote_average.median(),color='purple',linestyle='--')
mean_p=ax2.axhline(y=dm.popularity.median(),color='brown',linestyle='--')

plt.legend([vote, pop,mean_v,mean_p],
           ['vote_average median', 'popularity median','overall median of vote_average','overall median of popularity']
           ,bbox_to_anchor=(1.4,1.0),fontsize=12);

ax1.tick_params(axis='y', labelcolor='purple')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title('Counts of vote_average median and popularity median',fontsize=15);
ax1.set_xlabel('--------------> lower Genres counts (drama: largest / tv_movie: lowest)');
```

- Almost all the `vote_averge` value of the genres near `overall median of vote_averge` value, but the `horror` is lowest.
- The `popularity` has lots of difference in each genres. The `fantasy` and `animation` have the largest popularitis though they **rank 10** and **13** in overall counts.
- Although `drama` has the largest percentage, its pupularity is lower than overall median of popularity.
- The lowest pupularity is `documentary`,

```python
# calculate the descriptive statistics
g_max={'revenue':[],'ratio':[],'budget':[]}
g_min={'revenue':[],'ratio':[],'budget':[]}
g_mean={'revenue':[],'ratio':[],'budget':[]}
g_median={'revenue':[],'ratio':[],'budget':[]}
g_75={'revenue':[],'ratio':[],'budget':[]}

for i in Genres_sort.index:
    for j in ['ratio','revenue','budget']:
        tmp=dm.groupby(i)[j]
        try:
            g_max[j].append(tmp.max()[1])
        except:
            g_max[j].append(0)
        try:
            g_min[j].append(tmp.min()[1])
        except:
            g_min[j].append(0)
        try:
            g_mean[j].append(tmp.mean()[1])
        except:
            g_mean[j].append(0)
        try:
            g_median[j].append(tmp.median()[1])
        except:
            g_median[j].append(0)
        try:
            g_75[j].append(tmp.quantile(0.75)[1])
        except:
            g_75[j].append(0)
```

```python
def sub_plot_bar(ax1,ratio,revenue,budget,title,ratio_line=False):
    gr=ax1.bar(x,revenue,color=(1,0.5,0.6),width=w,align='center')

    ax2=ax1.twinx()
    gr2=ax2.bar(x+w,budget,color=(0.7,1,0.6),width=w,align='center')

    ax3=ax1.twinx()
    gr3=ax3.bar(x+2*w,ratio,color=(0.5,0.8,1),width=w,align='center')
    if ratio_line:
        line=ax3.axhline(y=1,color='k',linestyle='--')
        plt.legend([gr, gr2,gr3,line],['revenue', 'budget','ratio','ratio=1'],bbox_to_anchor=(1.4,1.0),fontsize=15);
    else:
        plt.legend([gr, gr2,gr3],['revenue', 'budget','ratio'],bbox_to_anchor=(1.4,1.0),fontsize=15);
    ax1.tick_params(axis='y', labelcolor='r')
    ax2.tick_params(axis='y', labelcolor='g')
    ax3.tick_params(axis='y', labelcolor='b')    
    

    ax3.spines["right"].set_position(("axes", 1.1))
    make_patch_spines_invisible(ax3)
    ax3.spines["right"].set_visible(True)
    

    ax1.set_ylabel("revenue",color='r')
    ax2.set_ylabel("budget",color='g')
    ax3.set_ylabel("ratio",color='b')


    plt.title(title,fontsize=12)
    

x=np.arange(len(Genres_sort.index))*2
w=0.5

plt.figure(figsize=(12,15))    

n=4
ax1=plt.subplot(n,1,1) 
ax2=plt.subplot(n,1,2) 
ax3=plt.subplot(n,1,3) 
ax4=plt.subplot(n,1,4) 
# ax5=plt.subplot(5,1,5)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=1)

sub_plot_bar(ax1,g_max['ratio'],g_max['revenue'],g_max['budget'],'Max')
sub_plot_bar(ax2,g_min['ratio'],g_min['revenue'],g_min['budget'],'Min')
sub_plot_bar(ax3,g_mean['ratio'],g_mean['revenue'],g_mean['budget'],'Mean')
sub_plot_bar(ax4,g_median['ratio'],g_median['revenue'],g_median['budget'],'Median',ratio_line=True)
# sub_plot_bar(ax5,g_75['ratio'],g_75['revenue'],g_75['budget'],'percentile = 75%',ratio_line=True)

ax1.set_xticks(x+w);ax1.set_xticklabels(new_genres_list,rotation=90);
ax2.set_xticks(x+w);ax2.set_xticklabels(new_genres_list,rotation=90);
ax3.set_xticks(x+w);ax3.set_xticklabels(new_genres_list,rotation=90);
ax4.set_xticks(x+w);ax4.set_xticklabels(new_genres_list,rotation=90);
# ax5.set_xticks(x+w);ax5.set_xticklabels(new_genres_list,rotation=90);
ax4.set_xlabel('--------------> lower Genres counts (drama: largest / tv_movie: lowest)');
```

- From the ratio, the `documentary` is more successful because it can earn many times than the budget (revenue >>> budget)
- The max ratio is in the `horror` and `mystery`, the min ratio is in the `documentary` and `foreign`
- The `drama` is rank 1, but its ratio is not the top one.
